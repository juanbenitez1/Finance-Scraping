```python
from numpy.core.fromnumeric import amin
import requests
from bs4 import BeautifulSoup
from datetime import date
import re 
import numpy as np 
import pandas as pd 
from IPython.display import clear_output
from time import sleep
from itertools import islice
import time
```

```python
def get_index(iterable, value, n):
    matches = (idx for idx, val in enumerate(iterable) if value in val)
    return next(islice(matches, n-1, n), None)
```

```python
secuencia = [1 + i*20 for i in range(0, (8261-21)//20 + 1)]

No = []
Company = []
Sector = []
Industry = []
Country = []
# Market_Cap = []
PE = []
PEG = []
PB = []
Dividend = []
EPS = []
ROA = []
ROE = []
DE = []
ProfitM = []
Beta = []
Earnings = []
Volume = []
# Price = []
Change = []
Ticker = []

for i in secuencia:
    main_url = 'https://finviz.com/screener.ashx?v=152&ft=2&r=7861&c=0,1,2,3,4,5,6,7,9,11,14,16,32,33,38,41,48,68,67,65,66&ft=2&r='+str(i)
    headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36"}
    main_response = requests.get(main_url, headers=headers)
    main_soup = BeautifulSoup(main_response.text,'html.parser')
    
    tabla = main_soup.find('table',class_='table-light')
    header = tabla.find('tr',align='center')
    header2 = header.find_all('td',class_='table-top cursor-pointer')
    cols = []

    for i in header2:
        col = i.text
        cols.append(col)
        
    rows = tabla.find_all('tr',valign='top')
    rows1 = []
    for i in rows:
        test = i.find_all('td',class_='screener-body-table-nw')
        temp = []
        for j in test:
            temp.append(j.text)
        rows1.append(temp)
        
    for i in rows1:
        No.append(i[0])
        Ticker.append(i[1])
        Company.append(i[2])
        Sector.append(i[3])
        Industry.append(i[4])
        Country.append(i[5])
        PE.append(i[7])
        PEG.append(i[8])
        PB.append(i[9])
        Dividend.append(i[10])
        EPS.append(i[11])
        ROA.append(i[12])
        ROE.append(i[13])
        DE.append(i[14])
        ProfitM.append(i[15])
        Beta.append(i[16])
        Earnings.append(i[17])
        Volume.append(i[18])
        Change.append(i[20])
```

```python
df = pd.DataFrame({'Simbol':Ticker,
                  'Company':Company,
                  'Sector':Sector,
                  'Industry':Industry,
                  'Country':Country,
                  'ROA':ROA,
                  'ROE':ROE,       
                  'Beta':Beta,       
                  'Debt_Equity':DE,                   
                  'P_E':PE,
                  'EPS_QQ':EPS,                
                  'PEG':PEG,
                  'Change':Change,                   
                  'Price_Book':PB,
                  'Payment_Date':Earnings,                   
                  'Dividend':Dividend,
                  'Profit_Margin':ProfitM,
                  'Volume':Volume
                  })

df = df.applymap(lambda x: np.nan if x == '-' else x)

cols_to_modify = ['ROA', 'ROE', 'Profit_Margin', 'Change',
                 'P_E', 'PEG', 'Price_Book', 'EPS_QQ',
                 'Debt_Equity', 'Beta']

for col in cols_to_modify:
    df[col] = df[col].astype(str)
    df[col] = df[col].str.replace('%', '')
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.head()
```
